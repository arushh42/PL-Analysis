```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
url_df2 = 'https://fbref.com/en/comps/9/2022-2023/2022-2023-Premier-League-Stats'
df2 = pd.read_html(url_df2)[0]
df2
```

```python
colors = ['lightgreen' if x >= 0 else 'lightcoral' for x in df2['xGD/90']]

plt.figure(figsize=(10, 6))
sns.barplot(x='Squad', y='xGD/90', data=df2, palette=colors)
plt.axhline(0, color='black', linewidth=1)  # Reference line at y=0
plt.title('Deviation of Actual Goals from Expected Goals')
plt.xlabel('Team')
plt.ylabel('Goals Deviation')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()


from matplotlib.table import Table
table_data = []
for i, (index, row) in enumerate(df2.iterrows()):
    table_data.append([row['Squad'], row['xGD/90'], row['Rk']])

table = plt.table(cellText=table_data, colLabels=['Team', 'Goals Deviation', 'League Position'], cellLoc='center', loc='bottom')
table.auto_set_font_size(False)
table.set_fontsize(10)

# Formatting the table
table.scale(1, 1.5)
plt.axis('off')  # Hide the axes
plt.show()
```

```python
url_df = 'https://fbref.com/en/comps/9/stats/Premier-League-Stats#all_stats_standard'
df = pd.read_html(url_df)[0]
df
```

```python
df.columns
```

```python
df.columns = [' '.join(col).strip() for col in df.columns]

df = df.reset_index(drop=True)
df.head()

new_columns = []
for col in df.columns:
  if 'level_0' in col:
      new_col = col.split()[-1]  # takes the last name
  else:
      new_col = col
  new_columns.append(new_col)

# rename columns
df.columns = new_columns
df = df.fillna(0)

df.head()
```

```python
plt.figure(figsize=(12, 6))

# Using Seaborn's barplot
sns.barplot(x='Squad', y='Performance CrdY', data=df, color='yellow')

# Adding labels and adjusting rotation for better readability
plt.xlabel('Squad')
plt.ylabel('Yellow Cards')
plt.title('Yellow Cards Distribution by Squad')
plt.xticks(rotation=45, ha='right')

plt.show()
```
